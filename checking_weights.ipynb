```python
df = pd.DataFrame()
diff = [ott.layers[3].get_weights()[0][j,:] - stj.layers[3].get_weights()[0][j,:] for j in range(stj.layers[3].get_weights()[0].shape[0])]

weights = [np.mean(ott.layers[3].get_weights()[0][:,:], axis=0), np.mean(stj.layers[3].get_weights()[0][:,:], axis=0), np.mean(diff, axis=0)]
print(np.shape(weights))
weights = np.array(weights)
weights = weights.reshape(weights.shape[1], weights.shape[0])
print(np.shape(weights))
vmin = np.min(weights)
vmax = np.max(weights)

fig = plt.figure(figsize=(5,20))

for i, k in enumerate(range(0,weights.shape[0],32)):
	ax = plt.subplot(int(weights.shape[0]/32),int(weights.shape[0]/32),i+1)
	ax.set_axis_off()
	im = ax.imshow(weights[k:(k+32),:], cmap='bwr', vmin=vmin, vmax=vmax)
	plt.subplots_adjust(wspace=0, hspace=0)

fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.8,
					wspace=0.02, hspace=0.02)

# cb_ax = fig.add_axes([0.83, 0.1, 0.02, 0.8])
# cb_ax = fig.add_axes([0.1, 0.1, 0.83, 0.05])
# cbar = fig.colorbar(im, cax=cb_ax, orientation='horizontal')
cbar = plt.colorbar(im)

plt.tight_layout()

plt.show()
```

```python
diff = [ott.layers[0].get_weights()[0][:,:,0,j] - stj.layers[0].get_weights()[0][:,:,0,j] for j in range(stj.layers[0].get_weights()[0].shape[3])]

weights = [ott.layers[0].get_weights()[0][:,:,0,:], stj.layers[0].get_weights()[0][:,:,0,:], diff]

titles = ['OTT', 'STJ', 'OTT-STJ']

vmin = np.min(weights)
vmax = np.max(weights)

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10,7))

for i, (ax, title) in enumerate(zip(axes.flat, titles)):
	ax.set_axis_off()
	ax.set_title(title)
	im = ax.imshow(weights[i].T, cmap='bwr',
				vmin=vmin, vmax=vmax)

fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.8,
					wspace=0.02, hspace=0.02)

# cb_ax = fig.add_axes([0.83, 0.1, 0.02, 0.8])
cb_ax = fig.add_axes([0.1, 0.1, 0.83, 0.05])
cbar = fig.colorbar(im, cax=cb_ax, orientation='horizontal')
# cbar = plt.colorbar(im, orientation='horizontal')

plt.tight_layout()

plt.show()
```
